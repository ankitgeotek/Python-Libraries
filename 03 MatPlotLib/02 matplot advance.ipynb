```python
#dataset
delivery=pd.read_csv(r"A:\Work Docs\DATA Set\IPL_Ball_by_Ball_2008_2022.csv")
delivery.sample(4)
```

```python
delivery['ballnumber'].unique()
```

```python
# we want only 6 ball of the over (here extra balls are also like noball, wide)
# also filtering over in which 6 are hit
temp_df=delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6]))&(delivery['batsman_run']==6)]
temp_df.head(2)
```

```python
grid=temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
# ploting grid by I am Show
plt.figure(figsize=(20,10))
p=plt.imshow(grid)

plt.xticks(np.arange(0,6), list(range(1,7)))
plt.yticks(delivery['overs'].unique(), list(range(1,21)))

plt.colorbar(p)
```
