```python
#label='xyz', plt.legend()
# lablel is legend and we should write plt.legent to see the legend
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dashed', linewidth=1, marker='d', markersize=5,label='Kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#ce0036',linestyle='dotted', linewidth=3,marker='*',markersize=10,label='Rohit')



plt.title('Rohit Sharma vs Virat Kohli career comparision')
plt.xlabel("IPL Season")
plt.ylabel('Runs Scored')

plt.legend(loc='center left')
# loc; supported values are 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'
```

# Bar Chart



* Numerical vs Catergorical Data
* Bivariante Analysis
* Use case- Aggregate Analysis of groups

```python
#Simple Bar Cart
Children=[10,20,50,25,15]
colors=['red','gree','pink','yellow','orange']
plt.bar(colors,Children,color='black')
plt.xlabel('Color')
plt.ylabel('No of childer')
```

```python
#Horizontal bar- barh, generally when no. of categories increases
plt.barh(colors,Children,color='black')
plt.xlabel('Color')
plt.ylabel('No of childer')
```

```python
#olor and labels
df=pd.read_csv(r"A:\Work Docs\DATA Set\batsman_season_record.csv")
df
```

```python
plt.bar(df['batsman'],df['2015'])
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2)
```

```python
# Multiple BarChart

plt.bar(np.arange(df.shape[0])-0.2,df['2015'],width=0.2,color='blue')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0])+0.2,df['2017'],width=0.2,color='green')

plt.xticks(np.arange(df.shape[0]),df['batsman'])
```

```python
#overlaping of x label
Children=[10,20,50,25,15]
colors=['red red red red red red','green green green green ','pink pink pink pink','yellowyellowyellow','orangeorangeorange']

plt.bar(colors,Children,color='black')
plt.xlabel('Color')
plt.ylabel('No of childer')
```

```python
#overlaping of x label
Children=[10,20,50,25,15]
colors=['red red red red red red','green green green green ','pink pink pink pink','yellowyellowyellow','orangeorangeorange']

plt.bar(colors,Children,color='black')
plt.xlabel('Color')
plt.ylabel('No of childer')
plt.xticks(rotation='vertical')
```

Stagged Bar Chart

```python

plt.bar(df['batsman'],df['2015'],color='green',label='2015')
plt.bar(df['batsman'],df['2016'],bottom=df['2015'],color='black',label='2016')
plt.bar(df['batsman'],df['2017'],bottom=(df['2016']+df['2015']),color='orange',label='2017')

plt.legend()
```

# Pie Chart



* Univarient/Bivarient Analysis
* Categorical vs numerical
* Use case- To find Contribution on a standard scale

```python
#Simple data
marks=[23,25,45,8,40,10]
subject=['Science','hindi','English','Social','Art','Physical']
plt.pie(data,labels=subject)
plt.show()
```

```python
# Adding Percentage perameter
plt.pie(data,labels=subject,autopct='%0.1f%%')
plt.show()
```

```python
#dataset
gayle=pd.read_csv(r"A:\Work Docs\DATA Set\gayle-175.csv")
gayle
```

```python
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
#percentage and color
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct='%0.1f%%',colors=['blue','green','yellow','orange','pink','black'])
plt.show()
```

```python
#Extracting pie-explored
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct='%0.1f%%',explode=[0,0.1,0,0,0,0])
plt.show()
```

```python
#shadow
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct='%0.1f%%',shadow=True)
plt.show()
```
