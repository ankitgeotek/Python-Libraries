```python
temp_dataframe=pd.DataFrame(
    {
        '1st Point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd Point':[(-3,4),(-0,0),(2,2),(10,-10),(1,1)]
    }
)
```

```python
temp_dataframe
```

```python
def euclidean(row):
    pt_A=row['1st Point']
    pt_B=row['2nd Point']

    return ((pt_A[0]-pt_B[0]**2)-(pt_A[1]-pt_B[1]**2))
```

```python
temp_dataframe['distance']=temp_dataframe.apply(euclidean,axis=1)
temp_dataframe
```
