### Figure level function -> `catplot`

```python
#Numerical scatter plot (both column numerical)
```

```python
# axes level function
# strip plot (one column categorical other numerical)
sns.stripplot(data=tips,x='day',y='total_bill')


```

```python
# removing noise (jitter)
sns.stripplot(data=tips,x='day',y='total_bill',jitter=False)
```

```python
# figure level function
# using catplot

sns.catplot(data=tips,x='day',y='total_bill',kind='strip')
```

```python
# increasing noise 
# jitter=0.1 (default)
sns.catplot(data=tips,x='day',y='total_bill',kind='strip', jitter=0.25)
```

```python
#figure level
# swarmplot
#better represention about distribution of data over strip plot
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm')
```

```python
#Axes level
# swarmplot
#better represention about distribution of data over strip plot
sns.swarmplot(data=tips,x='day',y='total_bill')
```

```python
# hue
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex')
```

### Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile [Q1], median, third quartile [Q3] and maximum). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

```python
# Box plot
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
sns.boxplot(data=tips,x='sex',y='total_bill')
```

```python
# Using catplot
sns.catplot(data=tips,x='day' ,y='total_bill' ,kind='box')
```

```python
# hue
sns.catplot(data=tips,x='day' ,y='total_bill' ,hue='sex',kind='box')
```

```python
# single boxplot -> numerical col
sns.catplot(data=tips,y='total_bill',kind='box')
```

### Regression Plots

- regplot
- lmplot

In the simplest invocation, both functions draw a scatterplot of two variables, x and y, and then fit the regression model y ~ x and plot the resulting regression line and a 95% confidence interval for that regression.

```python
tips
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
# regplot is axes level plot of regression plot
# hue parameter is not available
sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
# lmplot is figure level plot of regression plot
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
# residplot
# it tells the error done by regrassion line
sns.residplot(data=tips, x='total_bill', y='tip')
```

### A second way to plot Facet plots -> FacetGrid

```python
# figure level -> relplot -> displot -> catplot -> lmplot

sns.catplot(data=tips,x='sex',y='total_bill',kind='violin',col='day',row='time')
```

```python
#Creating Facet Grid

f_grid=sns.FacetGrid(data=tips,col='day',row='time')
```

```python
#Creating Facet Grid
g=sns.FacetGrid(data=tips,col='day',row='time')

#filling facetgrid
g.map(sns.violinplot,'sex','total_bill')
```

```python
#Adding Legeng 
g=sns.FacetGrid(data=tips,col='day',row='time', hue='smoker')

g.map(sns.violinplot,'sex','total_bill')
g.add_legend()
```

### Plotting Pairwise Relationship (PairGrid Vs Pairplot)

```python
iris
```

```python
# pair plot

sns.pairplot(data=iris)

```

```python
# pair plot with hue

sns.pairplot(data=iris,hue='species')
```

```python
#pair grid
g=sns.PairGrid(data=iris)
```

```python

#pair grid
g=sns.PairGrid(data=iris, hue='species')
#g.map
g.map(sns.histplot)
```

```python
#pair grid
g=sns.PairGrid(data=iris,hue='species')
#g.map
g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag
#pair grid
g=sns.PairGrid(data=iris,hue='species')
#g.map
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
# map_diag -> map_offdiag
#pair grid
g=sns.PairGrid(data=iris,hue='species')
#g.map
g.map_diag(sns.boxplot)
g.map_offdiag(sns.scatterplot)
```

```python
# map_diag -> map_upper -> map_lower
# map_diag -> map_offdiag
#pair grid
g=sns.PairGrid(data=iris,hue='species')
#g.map
g.map_diag(sns.histplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.kdeplot)
```

```python
# vars
#choose on what attributes you want to see plot
g=sns.PairGrid(data=iris,hue='species',vars=['sepal_length','petal_length'])
#g.map
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```
