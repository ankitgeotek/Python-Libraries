### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
#figure level -> relplot
#axes level -> scatterplot, lineplot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
#Scatter Plot-> axis level function


# sns.scatterplot(data=tips,x='total_bill', y='tip')
sns.scatterplot(x=tips['total_bill'], y=tips['tip'])
```

```python
#hue, style, size parameter


sns.scatterplot(data=tips,x='total_bill', y='tip',hue='sex',style='time',size='size')
```

```python
#Figure level plot-> Using relplot (relation plot)        scatter Plot
sns.relplot(data=tips,x='total_bill',y='tip', kind='scatter')
```

```python
#Figure level plot-> Using relplot (relation plot)        line Plot
#line plot is prefered in time series data


gap=px.data.gapminder()
gap
```

```python
temp_df=gap[gap['country']=='India']
temp_df
```

```python
#Axis level fuction
sns.lineplot(data=temp_df,x='year', y='lifeExp')
```

```python
#Using relplot i.e-fugure leve
sns.relplot(data=temp_df,x='year',y='lifeExp',kind='line')
```

```python
#hue->style

temp_df2=gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df2
```

```python
#in figure level legend is outside the graph
sns.relplot(data=temp_df2,x='year',y='lifeExp',hue='country',kind='line',style='continent',size='continent')
```

```python
#in Axis level legend is outside the , as axis level have control over axis only and not outside
sns.lineplot(data=temp_df2,x='year',y='lifeExp',hue='country')
```

```python
#facet plot-> Plotting multiple plot on categories (by column and row basis)
#facet work with figure level only it will not work with axisplot(i.e-lineplot,scatterplot)


sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',row='sex')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='smoker',row='sex')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex',row='day')
```

```python
#facet not work with axis plots
sns.scatterplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex')
```

```python
#col wrap
#fixing maxinmum no. of column per row

sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=3)
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# figure level -> displot
# axes level -> histplot, kdeplot, rugplot
```

```python
#plotting univaiate histogram
sns.histplot(data=tips,x='total_bill')
```

```python
#figure plot
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
# fixing bin number
sns.displot(data=tips,x='total_bill',kind='hist',bins=2)
```

```python
#Histogram acting like a count plot with categorical data
sns.displot(data=tips,x='day',kind='hist')
```

```python
#hue parameter

sns.displot(data=tips,x='tip',kind='hist',hue='sex')
```

```python
#Element in hue parameter

sns.displot(data=tips,x='tip',kind='hist',hue='sex',element='step')
```

```python
titanic=sns.load_dataset('titanic')
titanic
```

```python
#Element=step

sns.displot(data=titanic,x='age',kind='hist',element='step',hue='survived')
```

```python
#faceting using col and row -> not work with axis plot(i.e histplot)

sns.displot(data=titanic,x='age',kind='hist',element='step',col='sex')
```

```python
# kdeplot
# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate

#on Y-axis it give density of probabilty of data
sns.kdeplot(data=tips,x='total_bill')
```

```python
#figure plot
sns.displot(data=tips,x='total_bill', kind='kde',hue='sex')
```

```python
# hue -> fill

sns.displot(data=tips,x='total_bill', kind='kde',hue='sex',fill=True)

```

```python
# Rugplot

# sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
```

```python
#rug plot does not have utility alone

sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tile the plot
# and then shows the count of observations within each rectangle with the fill color

sns.histplot(data=tips, x='total_bill', y='tip')
```

```python
# Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
# Heatmap
# Plot rectangular data as a color-encoded matrix
# axes level function
gap
```

```python
temp_df3=gap.pivot(index='country',columns='year',values='lifeExp')
temp_df3
```

```python
#heatmap does not have figure level graph 
#Heatmpa has only Axes level graph

sns.heatmap(data=temp_df3)
```

```python
gap['continent'].value_counts()
```

```python
#resizing axes level graph
plt.figure(figsize=(15,15))
sns.heatmap(data=temp_df3)
```

```python
# annot
temp_df4=gap[gap['continent']=='Europe'].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(15,15))
sns.heatmap(data=temp_df4, annot=True)
```

```python
# Changing color of heatmap
#cmap

plt.figure(figsize=(15,15))
sns.heatmap(data=temp_df4, annot=True,cmap='viridis')
```

```python
# linewidth
# for giving gap between countries

plt.figure(figsize=(15,15))
sns.heatmap(data=temp_df4, annot=True,cmap='viridis',linewidths=0.5)
```

```python
# Clustermap

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.

iris = px.data.iris()
iris
```

```python
sns.clustermap(data=iris.iloc[:,0:4])
```
